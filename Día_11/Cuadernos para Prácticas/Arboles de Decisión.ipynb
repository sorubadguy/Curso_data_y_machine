```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
df = sns.load_dataset('iris')
df

```

```python
df.species.unique()

```

```python
X = df.drop('species', axis=1)
X
```

```python
le = LabelEncoder()
especies = le.fit_transform(df.species)
especies


```

```python
y = especies
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
arbol = DecisionTreeClassifier()
```

```python
arbol.fit(X_entrena, y_entrena)
```

```python
arbol.score(X_prueba, y_prueba)
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol);
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol, filled=True);
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol, 
          class_names=['setosa', 'versicolor', 'virginica'],
          filled=True);
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol, 
          class_names=['setosa', 'versicolor', 'virginica'],
          feature_names=df.columns.tolist(),
          filled=True);
```
