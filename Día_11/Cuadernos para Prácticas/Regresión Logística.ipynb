```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
ruta = 'datos_seguro.csv'
df = pd.read_csv(ruta)
df.head()
```

```python
plt.scatter(df.edad, df.compra);
```

```python
X_entrea, X_prueba, y_entrena, y_prueba = train_test_split(df[['edad']], df['compra'], train_size=0.9)#El 90% de los datos se usan para entrenar y el 10% para probar
```

```python
X_entrea
```

```python
X_prueba
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_entrea, y_entrena)
```

```python
modelo.score(X_entrea, y_entrena)
```

```python
datos_nuevos = pd.DataFrame({'edad' : [20, 30, 40, 50, 60]})
datos_nuevos
```

```python
probabilidades = modelo.predict_proba(datos_nuevos)
probabilidades

```

```python
prob_compra = probabilidades[:, 1]
prob_compra
```

```python
plt.scatter(df.edad, df.compra)
plt.scatter(datos_nuevos['edad'], prob_compra, color='red');
```
