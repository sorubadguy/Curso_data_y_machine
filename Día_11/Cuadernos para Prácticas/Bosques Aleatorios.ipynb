```python
# Librerías conocidas
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Librerías nuevas
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
ruta = 'tarjetas_credito.csv'

df = pd.read_csv(ruta)
df.head()
```

```python
escala = MinMaxScaler(feature_range=(0,1))
normado = escala.fit_transform(df)
df_normado = pd.DataFrame(normado, columns=df.columns)
df_normado.head()
```

```python
X = df_normado.drop(columns=['Clase'], axis=1)
```

```python
y = df_normado['Clase']
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
forest = RandomForestClassifier()
forest.fit(X_entrena, y_entrena)
```

```python
forest.score(X_prueba, y_prueba)
```
