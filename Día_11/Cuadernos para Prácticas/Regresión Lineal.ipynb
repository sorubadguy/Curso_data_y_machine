```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Area"] = [2600,3000,3200,3600,4000]
df["Precio"] = [550000,565000,610000,680000,725000]
```

```python
plt.scatter(df['Area'], df['Precio']);
```

```python
X = df[['Area']]
```

```python
type(df[['Area']])
```

```python
type(df['Area'])
```

```python
y = df['Precio']
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
modelo.predict([[3300]])
```

```python
plt.plot(df['Area'],
         modelo.predict(X));
```

```python
plt.scatter(df['Area'], 
            df['Precio'])
plt.plot(df['Area'],
         modelo.predict(X));
```

```python
df = pd.DataFrame()
df['Edad']= [25,30,35,40,45]
df['Salario'] = [3000,3500,4000,4500,5000]
```

```python
plt.scatter(df['Edad'], df['Salario'], marker='o', color='blue')
plt.title('Relación entre Edad y Salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.show()
```

```python
X = df[['Edad']]
y = df['Salario']
```

```python
modelo.fit(X,y)
```

```python
plt.plot(df['Edad'],
         modelo.predict(X));
plt.title('Relación entre Edad y Salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.show()
```
