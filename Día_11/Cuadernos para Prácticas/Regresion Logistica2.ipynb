```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
datos = {
    'edad': [25, 30, 45, 22, 34, 40, 27, 31, 43, 50],
    'salario': [50000, 54000, 72000, 42000, 62000, 68000, 49000, 56000, 76000, 90000],
    'compra': [0, 0, 1, 0, 1, 1, 0, 0, 1, 1]
}

df = pd.DataFrame(datos)
df
```

```python
plt.scatter(df.edad, df.compra);
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(df[['edad']], df['compra'])
```

```python
X_entrena
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_entrena, y_entrena)
```

```python
modelo.score(X_prueba, y_prueba)
```

```python
datos_nuevos = pd.DataFrame({'edad': [25, 30, 45, 55]})
```

```python
probabilidades = modelo.predict_proba(datos_nuevos)
probabilidades[:, 1]

```

```python
plt.scatter(df.edad, df.compra)
plt.scatter(datos_nuevos['edad'], probabilidades[:, 1], color='red');
```
