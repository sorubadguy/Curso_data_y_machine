```python
x = np.array(range(1,11))
y = np.power(x,2)
```

```python
fig, ax = plt.subplots()
ax.stem(x,y)
ax.set_title('RelaciÃ³n entre NÃºmeros y sus Cuadrados')
ax.set_ylabel('Cuadrado del nÃºmero (x^2)')
ax.set_xlabel('NÃºmero (x)')
```

```python
y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]
y = np.vstack([y1, y2, y3])
x = [1,2,3,4,5]
colors = ['red', 'green', 'blue']
```

```python
fig, ax = plt.subplots()
ax.stackplot(x, y, colors= colors)
ax.legend(['Serie y1','Serie y2','Serie y3'], loc='upper left')
ax.set_title('DistribuciÃ³n acumulativa de valores por serie de tiempo')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valores Acumulados');
```

```python
# Generar los datos para cada muestra usando numpy.random.normal
np.random.seed(0) # Establecer la misma semilla para la generaciÃ³n de datos
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)
```

```python
fig, ax = plt.subplots()
ax.violinplot([data1, data2, data3])
ax.set_xticks([1,2,3], labels=['Muestra 1', 'Muestra 2', 'Muestra 3']);
```
