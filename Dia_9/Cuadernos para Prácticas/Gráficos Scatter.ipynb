```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'o')
```

```python
aleatorio = np.random.RandomState(0)
marcadores = ['o', 'x', '+', 'v', 's', 'd']
```

```python
for marcador in marcadores:
    plt.plot(aleatorio.rand(5), aleatorio.rand(5), marcador)
```

```python
plt.scatter(x, y)
```

```python
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.random(100), aleatorio.random(100), c= aleatorio.randn(100), s=aleatorio.randn(100)*1000, alpha=0.3, cmap='viridis')
plt.colorbar();
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.scatter(x1, y1, color='blue', marker='o', label="Serie 1")
plt.scatter(x2, y2, color='green', marker='^', label="Serie 2")
plt.legend()
```
