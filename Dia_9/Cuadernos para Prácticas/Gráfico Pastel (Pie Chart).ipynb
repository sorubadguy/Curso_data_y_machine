```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 60, 80, 20]
})
df
```

```python
plt.pie(df['cantidad']);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas']);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%');
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87CEEB', '#6495ED', '#000080', '#40D0E0']);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87CEEB', '#6495ED', '#000080', '#40D0E0'], shadow=True);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87CEEB', '#6495ED', '#000080', '#40D0E0'],shadow=True, explode=(0,0,0,0.5));
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['#87CEEB', '#6495ED', '#000080', '#40D0E0'],shadow=True, explode=(0,0,0,0.5))
plt.title("Frutas");
```

```python
df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

plt.pie(df['cantidades'],labels=df['frutas'], autopct='%1.1f%%', colors = ['red', 'yellow', 'darkred', 'orange'], shadow=True, explode=(0,0,0.1,0))
plt.title("Distribucin de frutas en el almacn")
```
