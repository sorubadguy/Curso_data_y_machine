2. **Fusión de Datos**: Crea un segundo DataFrame que contenga los datos del archivo **Datos_Ventas_Tienda2.csv** (también provisto en esta lección), y concaténalos para tener un solo dataFrame con toda la información.

```python
import pandas as pd
tienda1 = 'Datos_Ventas_Tienda.csv'
tienda2 = 'Datos_Ventas_Tienda2.csv'
```

```python
df_t1 = pd.read_csv(tienda1)
df_t1
```

```python
df_t2 = pd.read_csv(tienda2)
df_t2
```

```python
df_concatenados = pd.concat([df_t1, df_t2])
tipo = df_concatenados['Fecha'].head(1)
tipo
```

```python
df_concatenados.info()
```

```python
producto_mas_vendido = df_concatenados.sort_values(by = ['Cantidad', 'Total Venta'], ascending = False).head(1)
producto_mas_vendido
```

```python
mes_con_mas_ventas = df_concatenados.groupby(df_concatenados['Fecha'].month)
mes_con_mas_ventas
```
