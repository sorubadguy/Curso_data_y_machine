# Códigos de Estado para las Respuestas HTTP

Estos son los diferentes códigos de estado que puedes recibir cuando hacer una petición HTTP a través del método `get()`.

| Código | Significado                          | Descripción                                                                              |
|--------|--------------------------------------|------------------------------------------------------------------------------------------|
| 100    | Continuar                            | El servidor ha recibido los encabezados de la solicitud y el cliente debe proceder a enviar el cuerpo de la solicitud. |
| 101    | Protocolos Cambiados                 | El servidor está cambiando a un protocolo diferente, según lo solicitado por el cliente.   |
| 102    | Procesando                           | El servidor ha recibido la solicitud y está procesándola, pero aún no tiene una respuesta lista. |
| 103    | Pistas Tempranas                     | El servidor envía algunos encabezados anticipados antes de la respuesta final.           |
| 200    | OK                                   | La solicitud ha tenido éxito.                                                            |
| 201    | Creado                               | La solicitud ha tenido éxito y se ha creado un nuevo recurso.                            |
| 202    | Aceptado                             | La solicitud ha sido aceptada para su procesamiento, pero no se ha completado.           |
| 203    | Información No Autorizada            | La información devuelta ha sido modificada por un proxy.                                 |
| 204    | Sin Contenido                        | La solicitud ha tenido éxito, pero no hay contenido que devolver.                        |
| 205    | Restablecer Contenido                | La solicitud ha tenido éxito, pero el agente del usuario debe restablecer el documento visto. |
| 206    | Contenido Parcial                    | El servidor está entregando solo parte del recurso debido a un encabezado de rango enviado por el cliente. |
| 207    | Multi-Estado                         | Proporciona información de estado para varios recursos.                                 |
| 208    | Ya Reportado                         | Los miembros del conjunto de enlaces DAV ya han sido enumerados en una respuesta anterior. |
| 226    | IM Usado                             | El servidor ha cumplido una solicitud GET para el recurso y la respuesta es una representación del resultado de una o más manipulaciones de instancias aplicadas a la instancia actual. |
| 300    | Múltiples Opciones                   | Hay múltiples opciones para el recurso del cliente.                                      |
| 301    | Movido Permanentemente               | El recurso solicitado ha sido movido permanentemente a una nueva URL.                    |
| 302    | Encontrado                           | El recurso solicitado reside temporalmente en una URL diferente.                         |
| 303    | Ver Otros                            | El cliente debe realizar una nueva solicitud en una URL diferente.                       |
| 304    | No Modificado                        | El recurso no ha sido modificado desde la última solicitud.                              |
| 305    | Usar Proxy                           | El recurso solicitado debe accederse a través del proxy indicado.                        |
| 306    | Cambiar Proxy                        | Este código ya no se utiliza, pero está reservado.                                       |
| 307    | Redirección Temporal                 | El recurso solicitado reside temporalmente en una URL diferente.                         |
| 308    | Redirección Permanente               | El recurso solicitado ha sido movido permanentemente a una nueva URL.                    |
| 400    | Solicitud Incorrecta                 | El servidor no puede o no procesará la solicitud debido a un error del cliente.          |
| 401    | No Autorizado                        | La solicitud requiere autenticación del usuario.                                         |
| 402    | Pago Requerido                       | Esta respuesta se reserva para uso futuro.                                               |
| 403    | Prohibido                            | El servidor ha entendido la solicitud, pero se niega a autorizarla.                      |
| 404    | No Encontrado                        | El servidor no puede encontrar el recurso solicitado.                                    |
| 405    | Método No Permitido                  | El método de solicitud no está permitido para el recurso solicitado.                     |
| 406    | No Aceptable                         | El recurso solicitado solo puede generar contenido no aceptable según los encabezados Accept enviados en la solicitud. |
| 407    | Autenticación de Proxy Requerida     | El cliente debe autenticarse con el proxy.                                               |
| 408    | Tiempo de Solicitud Agotado          | El servidor agotó el tiempo de espera mientras esperaba la solicitud del cliente.        |
| 409    | Conflicto                            | La solicitud no se puede completar debido a un conflicto en el estado del recurso.       |
| 410    | Desaparecido                         | El recurso solicitado ya no está disponible y no lo estará de nuevo.                     |
| 411    | Longitud Requerida                   | El servidor rechaza la solicitud porque no se ha definido el encabezado Content-Length.  |
| 412    | Fallo de Precondición                | El servidor no cumple una de las precondiciones que el solicitante puso en los encabezados. |
| 413    | Entidad de Solicitud Demasiado Grande| La entidad de la solicitud es más grande de lo que el servidor puede manejar.            |
| 414    | URI Demasiado Larga                  | La URI de la solicitud es demasiado larga para que el servidor la procese.               |
| 415    | Tipo de Medio No Soportado           | El tipo de medio de la solicitud no es soportado por el servidor.                        |
| 416    | Rango No Satisfactorio               | El rango especificado por el encabezado Range del cliente no puede ser cumplido.         |
| 417    | Fallo en la Expectativa              | La expectativa indicada por el encabezado Expect del cliente no puede ser cumplida por el servidor. |
| 418    | Soy una tetera                       | El servidor se niega a preparar café porque es una tetera.                              |
| 421    | Solicitud Mal Dirigida               | La solicitud fue dirigida a un servidor que no es capaz de producir una respuesta.       |
| 422    | Entidad No Procesable                | La solicitud estaba bien formada pero no se pudo procesar debido a errores semánticos.  |
| 423    | Bloqueado                            | El recurso al que se accede está bloqueado.                                              |
| 424    | Fallo de Dependencia                 | La solicitud falló debido a la falla de una solicitud previa.                           |
| 425    | Demasiado Temprano                   | El servidor no está dispuesto a arriesgarse a procesar una solicitud que podría repetirse. |
| 426    | Actualización Requerida              | El cliente debe actualizarse a un protocolo diferente.                                   |
| 428    | Precondición Requerida               | El servidor requiere que la solicitud sea condicional.                                   |
| 429    | Demasiadas Solicitudes               | El usuario ha enviado demasiadas solicitudes en un tiempo dado.                          |
| 431    | Campos de Encabezado de Solicitud Demasiado Grandes | El servidor no está dispuesto a procesar la solicitud porque los campos de encabezado son demasiado grandes. |
| 451    | No Disponible por Razones Legales    | El recurso solicitado no está disponible debido a razones legales.                       |
| 500    | Error Interno del Servidor           | El servidor encontró una condición inesperada que le impidió cumplir con la solicitud.  |
| 501    | No Implementado                      | El servidor no reconoce el método de solicitud o no tiene la capacidad para completarlo. |
| 502    | Puerta de Enlace Incorrecta          | El servidor, actuando como una puerta de enlace o proxy, recibió una respuesta inválida del servidor ascendente. |
| 503    | Servicio No Disponible               | El servidor no está disponible para manejar la solicitud debido a un mantenimiento o sobrecarga. |
| 504    | Tiempo de Espera de la Puerta de Enlace Agotado | El servidor, actuando como una puerta de enlace o proxy, no recibió una respuesta a tiempo del servidor ascendente. |
| 505    | Versión HTTP No Soportada            | El servidor no soporta la versión del protocolo HTTP utilizada en la solicitud.          |
| 506    | La Variante También Negocia          | El servidor tiene un error de configuración interna: el recurso negociado de contenido está configurado para usarse a sí mismo en la negociación. |
| 507    | Almacenamiento Insuficiente          | El servidor no puede almacenar la representación necesaria para completar la solicitud.  |
| 508    | Bucle Detectado                      | El servidor detectó un bucle infinito mientras procesaba la solicitud.                   |
| 510    | No Extendido                         | El servidor requiere más extensiones para cumplir con la solicitud.                     |
| 511    | Autenticación de Red Requerida       | El cliente necesita autenticarse para obtener acceso a la red.                           |
