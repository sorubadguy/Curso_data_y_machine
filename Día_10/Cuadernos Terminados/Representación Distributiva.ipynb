# Representación Distributiva

En esta lección nos vamos a abocar al segundo conjunto de gráficos, que como vamos a ver, están contenidos dentro de la función `displot()`, la cual se especializa en **Representación Distributiva**, que es otro concepto de la estadística que vamos a explicar ahora mismo.


#### Introducción
Dentro del análisis estadístico, entender las relaciones entre variables (que es lo que vimos en la lección anterior), es tan crucial como conocer la distribución de tus datos, es decir, **cómo se distribuyen la información que dispones**.

La **representación distributiva** se refiere al proceso de visualizar cómo se distribuyen los valores de una o más variables dentro de un conjunto de datos.

Cuando hablamos de distribución, básicamente estamos diciendo cuántos casos tenemos en diferentes grupos. Por ejemplo, tengo un grupo de **10 personas** que se distribuyen:
* por **género** en:
    * 6 hombres
    * 4 mujeres
* por **edad** en:
    * 3 personas de 0 a 10 años
    * 5 personas entre 11 y 20
    * 2 personas entre 21 y 30
* por **estado civil** en:
    * 4 solteros
    * 5 casados
    * 1 divorciado


Entonces la representación distributiva se refiere a **cómo representamos la distribución de nuestros datos en grupos**, y en esta lección, nos vamos a concentrar en cómo Seaborn facilita la representación de distribuciones de una manera intuitiva y muy poderosa.


#### Explorando Distribuciones con Seaborn
Seaborn nos brinda una función capital para representar distribuciones que es la función `displot()`, que como vemos en esta imagen, incluye las opciones de representar **histogramas**, **gráficos kde**, **gráficos ecdf** y **rugplots**. Estos nombres puede que aún no sean muy significativos para ti, pero ya lo serán.

![](funciones_seaborn.png)

Vamos a importar Seaborn a nuestro cuaderno y a cargar una base de datos de ejemplos compuesta por información sobre **pingüinos**.

```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos.head()
```

### Histogramas con displot()

Ahora voy a representar la distribusión de la columna `body_mass_g` (masa corporal) usando `displot()`. Al no especificar ningún tipo en particular, automáticamente `displot()` va a producir un **histograma**:

```python
sns.displot(data=pinguinos,
           x="body_mass_g");
```

En este gráfico resultante tenemos un **histograma** que muestra cómo se distribuyen los registros de **masa corporal** de los pingüinos en diferentes agrupaciones.

### KDE en displot()

Dijimos que los `displot()` pueden contener otros tipos de gráfico, como los `kde`, a los que agregamos justamente usando el parámetro `kde` como **True**:


```python
sns.displot(data=pinguinos,
           x="body_mass_g",
           kde=True);
```

**KDE** significa **Estimación de la Densidad de Kernel** en inglés, y es una técnica utilizada para visualizar la distribución de manera suave y continua.

A diferencia de los histogramas, que dividen los datos en bins o contenedores para mostrar la frecuencia de los valores dentro de cada bin, los gráficos **KDE** trazan una curva que representa la densidad de probabilidad estimada de la variable. Esto nos proporciona una vista más limpia y continua de la distribución, lo cual es útil para identificar la forma de la distribución, como picos, asimetría y dispersión.

Y no olvidemos que también disponemos del atributo `col`, para hacer comparaciones entre distintos valores de algún otro atributo, por ejemplo, en este caso, las **islas** donde se encuentran estos pingüinos.

```python
sns.displot(data=pinguinos,
           x="body_mass_g",
           kde=True,
           col="island");
```

Entonces así ya hemos visto cómo graficar tanto **histogramas** como **kdeplots** dentro de `displot()`, por lo que nos faltaría ver los gráficos **rugplot** y **ecdfplot** que también están incluidos en `displot()`.


### Rugplot en displot()

Podemos agregar un **rugplot** en nuestros gráficos madificando el valor del atributo `rug` como **True**:

```python
sns.displot(data=pinguinos,
           x="flipper_length_mm",
           kind="kde",
           rug=True);
```

Un **rugplot** es una forma compacta de visualizar la distribución de los datos. Nos muestra pequeñas marcas (o "flecos") para cada punto de dato a lo largo de un eje.

Como puedes ver, es útil para ver la densidad de los datos.

Al haber hecho un **kde** para generar una estimación de densidad kernel de la longitud de las aletas de los pingüinos, se puede ver la curva que representa la distribución de esos valores. Y el **rugplot** que hemos agregado, nos proporciona unas marcas individuales para cada observación, ofreciendo una doble perspectiva de la densidad y la distribución de los datos.



### ECDF en displot()

Y ahora veamos el gráfico **ecdf**, al que cargamos como valor del atributo `kind`:

```python
sns.displot(data=pinguinos,
           x="flipper_length_mm",
           kind="ecdf");
```

El **ECDF** es un gráfico de la función de **distribución acumulativa empírica**, y nos muestra la proporción de observaciones que caen por debajo de cada valor, lo que brinda una vista completa de la distribución de los datos. En este caso tenemos un **ECDFplot** de la longitud de las aletas de los pingüinos, en el que se van agregando los casos a medida que subimos en los valores del **eje x**, hasta llegar a tener todos los casos acumulados al final.
