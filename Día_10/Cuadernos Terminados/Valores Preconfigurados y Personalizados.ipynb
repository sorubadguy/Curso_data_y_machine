# Valores Preconfigurados vs. Personalizados

Como hemos visto el día de hoy, Seaborn cuenta con **funciones de alto** y de **bajo nivel**.

Con las funciones de **alto nivel**, Seaborn crea gráficos completos con **una sola llamada a la función**. Cuando es posible, estas funciones van a agregar automáticamente etiquetas de ejes informativas y leyendas para  explicar los mapeos semánticos en el gráfico.

En esos casos, Seaborn, que es muy inteligente e intuitivo, también va a elegir los valores predeterminados para los parámetros de esas funciones, basándose en las características de los datos. Por ejemplo, los mapeos de color que hemos visto hasta ahora usaron tonos distintos (azul, naranja y a veces verde) para representar diferentes niveles de las variables categóricas asignadas al parámetro `hue`. Y en el caso de las variables numéricas, algunas funciones lo que hacen es cambiar esos colores planos a un gradiente continuo.

Ejecutemos el siguiente código, que va a generar el dataset de los pingüinos, y luego ejecutará un `relplot()` muy simple que muestra un **histograma**:

```python
import seaborn as sns
import matplotlib as plt
```

```python
pinguinos = sns.load_dataset("penguins")
```

```python
sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g");
```

Este gráfico se ve muy bien, y eso es porque `relplot()` ha tomado muchas decisiones gráficas por nosotros, basándose en todo lo que sabe.

Esto es genial para la etapa en que estamos jugando y explorando nuestros datos, pero cuando estés listo para compartir o publicar tu trabajo, probablemente vas a querer pulir la figura más allá de lo que se logra con los valores predeterminados. 

En esos casos, Seaborn permite **varios niveles de personalización**. Entre otros recursos Seaborn tiene predefinidos múltiples **temas** incorporados (que son como colecciones de estilos visuales), que se aplican a todas las figuras al mismo tiempo. Y no solo eso, sino que una vez que hayas creado un gráfico, sus propiedades pueden modificarse tanto a través de Seaborn como descendiendo al nivel de matplotlib para un ajuste detallado.

Si establezco el **tema** ahora, esto afectará la forma en que se verán todos mis gráficos de ahora en adelante. Lo haré utilizando `set_theme()`:

```python
sns.set_theme(style="dark")

sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g");
```

He establecido el valor `dark`, que es un **tema** predefinido, y de ahora en más, todos los gráficos que escriba, se mostrarán con este estilo de fondo gris.

Ahora, voy a copiar nuevamente el código del gráfico anterior, y vamos a ver muchas de las cosas que podemos hacer si queremos personalizarlo en un nivel elevado. Entonces dentro del `relplot()`, voy a agregar algunos parámetros más:

```python
sns.set_theme(style="dark")

sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette="crest",
           marker="x",
           s=100);
```

Como has visto, he incorporado 3 nuevos parámetros, que son:
* `palette="crest"`: establece una paleta de colores llamada `crest`. Puedes encontrar los nombres de las diferentes paletas en la documentación
* `marker="x"`: establece el formato de los puntos de datos
* `s=100`: establece el tamaño de los puntos

Lo importante que quiero que aprendas aquí no es necesariamente esos parámetros en partivcular, sino mostrarte un ejemplo en el que voy cambiando parámetros que personalizan a un gráfico de **alto nivel**. De esta manera puedes tomar decisiones más personales desde el punto de vista estético sobre cómo quieres que se vean tus gráficos a la hora de compartirlos

Aquí vas a aplicar criterios no solamente de **preferencia estética**, sino que también **criterios estratégicos** para lo que tu creas que es mejor, en cuanto a lograr que la información sea más fácil de comprender.

**Un consejo**: en esta parte no te vuelvas ambicioso. No es cuestión de aplicar todos los recursos posibles, o de personalizar todos los parámetros disponibles solo porque sí, sino por el contrario, hacer que esos recursos nos ayuden a simplificar la comprensión. Muchas veces en el diseño gráfico "menos es más".

Ahora voy a ir trabajando con los elementos particulares del gráfico a través de la nomenclatura de punto, y para eso voy a guardar a nuestro gráfico en una variable `g`, de modo que pueda así invocar a sus elementos por separado.

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette="crest",
           marker="x",
           s=100)
```

### Personalizar etiquetas de eje

Primero voy a personalizar las etiquetas de los ejes (las que contienen los nombres de las variables), y para eso voy a usar la función `set_axis_labels()`.

Las primeras dos propiedades serán los títulos que quiero asignar a las etiquetas de los ejes **x** e **y** respectivamente, y la propiedad `labelpad` para un espaciado adicional que evite que las etiquetas se peguen demasiado a los ejes.

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette="crest",
           marker="x",
           s=100)
g.set_axis_labels("Largo del Pico (mm)",
                "Grosor del Pico (mm)",
                labelpad=10);
```

### Personalizar el título de la leyenda

Del mismo modo puedo modificar el **título de la leyenda**, si quiero que diga otra cosa, usando la función `set_title` perteneciente al objeto `legend`:

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette="crest",
           marker="x",
           s=100)
g.set_axis_labels("Largo del Pico (mm)",
                "Grosor del Pico (mm)",
                labelpad=10)
g.legend.set_title("Masa\nCorporal (g)");
```

### Ajustar el tamaño de la figura

Ahora, a mi gusto, el gráfico se está viendo un poco apretado, porque es demasiado angosto para la cantidad de información que contiene. Creo que si modificamos la proporción de la figura vamos a tener una visualización más clara.

Para hacer esto llamaré al objeto `figure`, que contiene una función llamada `set_size_inches()`. Esta función nos pide que pasemos en pulgadas una tupla con los valores para *ancho* y *alto* respectivamente.

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette="crest",
           marker="x",
           s=100)
g.set_axis_labels("Largo del Pico (mm)",
                "Grosor del Pico (mm)",
                labelpad=10)
g.legend.set_title("Masa\nCorporal (g)")
g.figure.set_size_inches(6.5, 4.5);
```

### Aumentar los márgenes dentro del gráfico
Otro aspecto que me gustaría modificar, es que los puntos dentro del gráfico están muy cercanos a los bordes. Lo que quiero hacer es **ampliar el margen** alrededor del gráfico para que los datos queden más separados y de ese modo la visualización sea más limpia.

Para lograr este objetivo, he encontrado que el objeto `ax` (eje), contiene una función llamada `margins()`, que aumenta los márgenes alrededor del área de dibujo dentro del gráfico en un porcentaje establecido por nosotros.

```python
g = sns.relplot(data=pinguinos,
            x="bill_length_mm",
            y="bill_depth_mm",
            hue="body_mass_g",
           palette="crest",
           marker="x",
           s=100)
g.set_axis_labels("Largo del Pico (mm)",
                "Grosor del Pico (mm)",
                labelpad=10)
g.legend.set_title("Masa\nCorporal (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15);
```

Ahora el resultado me encanta, creo que es un buen gráfico para mostrar o publicar, y si bien me ha tomado más trabajo que cuando usamos funciones de alto nivel sin ninguna especificación extra, cada una de estas líneas contribuye a la creación de un gráfico de dispersión altamente personalizado y estéticamente agradable, facilitando la interpretación de los datos que quiero compartir.
