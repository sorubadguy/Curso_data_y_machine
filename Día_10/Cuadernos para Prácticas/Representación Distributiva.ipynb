```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset('penguins')
pinguinos.head()
```

```python
sns.displot(data=pinguinos,
            x = 'body_mass_g');
```

```python
sns.displot(data=pinguinos,
            x = 'body_mass_g',
            kde = True);
```

```python
sns.displot(data=pinguinos,
            x = 'body_mass_g',
            kde = True,
            col='island');
```

```python
sns.displot(data=pinguinos,
            x='flipper_length_mm',
            kind='kde',
            rug=True);
```

```python
sns.displot(data=pinguinos,
            x='flipper_length_mm',
            kind='ecdf');
```
