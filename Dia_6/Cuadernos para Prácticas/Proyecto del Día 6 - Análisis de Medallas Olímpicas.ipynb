```python
import pandas as pd
```

```python
medallas_df = pd.read_csv("medallas.csv")
medallas_df
```

```python
medallas_df.isnull().sum()
```

```python
medallas_df['Total'].std()
```

```python
correccion_medallas = {"Oro" : 0, "Plata" : 0, "Bronce" : 0}
```

```python
medallas_df = medallas_df.fillna(correccion_medallas)
medallas_df
```

```python
medallas_df['Oro'] = medallas_df['Oro'].astype(int)
medallas_df['Plata'] = medallas_df['Plata'].astype(int)
medallas_df['Bronce'] = medallas_df['Bronce'].astype(int)
medallas_df['Total'] = medallas_df['Total'].astype(int)
```

```python
dir(medallas_df)
```

```python
medallas = {}
for indice_fila, fila in medallas_df.sort_values('Oro', ascending = False).head(3).iterrows():
    medallas[fila['Pais']] = fila['Oro']

medallas_oro = pd.Series(medallas)
medallas_oro
```

```python
medallas = {}
for indice_fila, fila in medallas_df.sort_values('Total', ascending = False).iterrows():
    if(int(fila['Total']) > 10):
        medallas[fila['Pais']] = fila['Total']

medallas_totales = pd.Series(medallas)
medallas_totales
```
